```python
# bins = 범위를 나타냄 [1,2,3,4] 인 경우 1~2, 2~3, 3~4 범위를 나타냄
import matplotlib.pyplot as plt

data = [1,2,3,4,5,7,8,9,10]

plt.hist(data, rwidth = .98) # rwidth = 범위를 나타냄
plt.show()
```

```python
data = [10, 1, 2, 3,4, 5, 7, 8, 9, 1]
plt.hist(data, bins = 3, rwidth=.98) # rwidth : 구간의 너비 0.98이니까 오른쪽으로 2%씩 빠짐
print(f'하나의 구간 길이 : {(max (data) - min(data)) / 3 : .2f}')
plt.show()
```

```python
import numpy as np

dice = np.random.choice(range(1,7), 10) # 1~6 사이의 숫자를 10번 랜덤으로 선택

plt.figure(figsize=(2,1), dpi = 200) # 그래프 크기 설정 dpi : 해상도
plt.hist(dice, rwidth = .98, bins=[.5,1.5,2.5,3.5,4.5,5.5,6.5])
plt.show()
```

```python
dice = np.random.choice(range(1,7), 10000000, p = [0.15, 0.25, 0.3, 0.1, 0.1, 0.1]) # 1~6 사이의 숫자를 10번 랜덤으로 선택

plt.figure(figsize=(3, 2), dpi = 150) # 그래프 크기 설정 dpi : 해상도
plt.hist(dice, rwidth = .98, bins=np.arange(.5, 7)) 
plt.xticks(range(1,7)) # x축 눈금 설정
plt.show()
```

```python
a = np.arange(-np.pi, np.pi, np.pi/100)

plt.plot(a, np.sin(a), label='sin(a)')
plt.plot(a, np.cos(a), label='cos(a)')
plt.plot(a + np.pi / 2, np.sin(a) , label = ' ... sin(a)') # np.pi만큼 x축으로 이동한 sin(a) 그래프
plt.legend()
plt.show()
```

```python
a = []
b = []
c = []

for i in range(0,50,2):
    a.append(i / 10)
    b.append((i / 10 ) ** 2)
    c.append((i / 10 ) ** 3)

plt.plot(a, a, 'r--', a, b, 'bs', a, c, 'g^')
plt.show()
```

```python
# 마스크(Mask) 만들고 적용하기

a = np.arange(-5, 5)
print(a)

print(a < 0) # 논리형으로 출력
print(a[a < 0]) # 논리형에 해당하는 배열 값 출력
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)

mask1 = (abs(x) > 50)
mask2 = (abs(y) > 50)

x = x[mask1 + mask2]
y = y[mask1 + mask2]

plt.scatter(x, y, cmap="viridis", alpha=0.5)
plt.colorbar()
plt.show()
```
