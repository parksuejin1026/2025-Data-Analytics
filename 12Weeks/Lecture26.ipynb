```python
import csv
import numpy as np
import matplotlib.pyplot as plt

# 1. errors='ignore'를 추가하여 인코딩 에러 방지
f = open('age.csv', encoding='cp949', errors='ignore')
data = csv.reader(f)
next(data)

name = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요: ")
home = None  # 3. 변수 초기화

for row in data :
    if not row: continue # 빈 줄일 경우 건너뛰기
    
    row[2:] = map(lambda s : s.replace(',', ''), row[2:]) # ,를 제거
    if name in row[0] :
        # 2. dytpe -> dtype 오타 수정
        home = np.array(row[3:], dtype = int)
        break

f.close()

# 3. 데이터를 찾았을 때만 그래프 그리기
if home is not None:
    plt.rc('font', family = 'Malgun Gothic')
    plt.plot(home)
    plt.title(f'{name} 인구 구조')
    plt.grid(True)
    plt.xlabel('연령')
    plt.ylabel('인구수')
    plt.show()
else:
    print(f"'{name}' 지역을 찾을 수 없습니다.")
```
