```python
# Numpy 개요와 사용
import numpy as np

np.array([1,2,3])

a = np.ones(2) # 2개의 1.로 이루어진 배열 ones default는 float64임
print(a) # [1. 1.]
print(a.dtype) # float64
```

```python
np.ones(2, dtype=np.int64) # 2개의 1로 이루어진 배열, int64로 이루어진 배열

```

```python
print(np.arange(4)) # 0부터 4까지
print(np.arange(2, 9, 2)) # 2부터 9까지 2씩 증가
print(np.arange(1, 5, .7)) # .7은 소수점
```

```python
print(np.linspace(0, 10, 5)) # 0부터 10까지 5등분
print(np.linspace(0, 10, num = 5)) #
print(np.linspace(0, 10, num = 11))
```
