막대 그래프 (Bar Chart)

```python
import matplotlib.pyplot as plt # 그래프를 그리는 데 사용되는 라이브러리
import pandas as pd # 데이터프레임(표 형태)을 다루는데 사용되는 라이브러리

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

# 데이터 생성 : 도시와 인구 정보를 딕셔너리 혀앹로 정의
data = {'도시': ['서울','부산', '대구', '인천'],
        '인구': [970, 340, 240, 290]}
df = pd.DataFrame(data) # 딕셔너리를 Pandas 데이터프레임으로 변환

# 그래프 설정 및 그리기
plt.figure(figsize=(8,6)) # 그래프의 가로, 세로 크기 설정
# plt.bar(): 막대 그래프를 그리는 함수. x축에 '도시', y축에 '인구'를 표시
plt.bar(df['도시'], df['인구'], color=['skyblue', 'salmon', 'lightgreen', 'gold'])
plt.title('도시별 인구 현황') # 그래프의 제목 설정
plt.xlabel('도시')
plt.ylabel('인구 (만 명)')
plt.grid(axis='y', linestyle='--', alpha=0.7) # y축에만 점선 그리드 표시
plt.show() # 그래프를 화면에 출력
```
