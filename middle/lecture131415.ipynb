```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

result = []

for row in data:
    if row[4] != '': # 최고 기온 데이터가 존재한다면
        if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '25'):  # 12월에 해당하는 값이라면 이렇게 쓰는 이유는 2012-12-25 이런 형식으로 되어있기 때문이다.
            result.append(float(row[4])) # 리스트 result에 최고 기온 값을 추가

f.close() 
print(len(result))

plt.figure(figsize=(10,2))
plt.plot(result, 'r')
plt.xlabel('Day')
plt.ylabel('The highest temparature')
plt.show()

s = 'Happy New Year!'
print(s.split())
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

result = []
for row in data :
    if row[-1] != '':
        if '12-25' in row[0]:
            result.append(float(row[-1]))

plt.figure(figsize=(10,2))
plt.plot(result, 'hotpink')
plt.xlabel('Day (December 25)')
plt.ylabel('The Highest Temperature')
plt.show()
```
