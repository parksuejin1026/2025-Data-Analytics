```python
fig, axes = plt.subplots(2, 2)
np.random.seed(0)
x = np.random.rand(5)
axes[0,0].plot(x)
axes[0,0].set_title("axes 1")
axes[0,1].plot(x)
axes[0,1].set_title("axes 2")
axes[1,0].plot(x)
axes[1,0].set_title("axes 3")
axes[1,1].plot(x)
axes[1,1].set_title("axes 4")

plt.tight_layout()
plt.suptitle('Figure Title')
plt.show()
```

```python
fig = plt.figure(figsize=(7, 5))
data = np.random.randn(50)

ax2 = fig.add_subplot(2, 2, 2)
plt.plot(data, 'g--')
ax3 = fig.add_subplot(2, 2, 3)
plt.plot(data.cumsum(), 'r.-')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins=100, alpha=0.7, histtype='step', label='step')
plt.hist(b, bins=50, alpha=0.5,histtype='stepfilled', label='stepfilled')
plt.hist(c, bins=100, alpha=0.3, histtype='bar', label='bar')

plt.legend()
plt.show()
```

```python
fig , ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', 'brown', 'orange']
bar_colors = ['red', 'blue', 'brown', 'orange']
ax.bar(fruits, counts, label=bar_labels, color=bar_colors)
#    x    ,    y  

ax.set_ylabel('fruit supply')
ax.set_title('fruit supply by kind and color')
ax.legend(title = 'fruit color')

plt.show()
```
