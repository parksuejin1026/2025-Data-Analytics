```python
import pandas as pd
import numpy as np  
```

```python
a = pd.Series([3, 5.6, 7, 9], name = 'basic')
a
```

```python
print(a.name)
print(a.values)
print(a.index)
```

```python
a = pd.Series([3, -5.6, 7, 9], index = list('abcd'), name = 'basic')
a
```

```python
print(a.name)
print(a.values)
print(a.index)
```

```python
df = pd.DataFrame(
    {
        "A" : 1.0, # 
        "B" : np.arange(0,4, .5), # 0부터 4까지 0.5씩 증가
        "C" : pd.Timestamp("20250102"), # 날짜
        "D" : np.array([3] * 8, dtype = "int32"), # 3으로 구성된 배열
        "F" : "foo",
    }, index = pd.date_range("20250102", periods = 8) # 날짜 범위
)
df
```

```python
df.index # index 반환
```

```python
df.columns # 컬럼 확인
```

```python
df.dtypes # 데이터 타입 확인
```
