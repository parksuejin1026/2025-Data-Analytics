```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data) # 헤더를 제외한 데이터를 가져옴
next(data) # 헤더를 제외한 데이터를 가져옴
print(header)

result = []
for row in data: 
    for i in range(4, 52): 
        row[i] = row[i].replace(',', "")
    row[4:] = map(int, row[4:]) # 4번째 인덱스부터 52번째 인덱스까지의 데이터를 int로 변환
    result.append(row[10]) # 10 번째 인덱스 데이터를 리스트에 저장 

f.close()

result.sort() # 오름차순으로 정렬
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize = (10, 3))
plt.bar(range(len(result)), result)
plt.xlabel("역")
plt.ylabel("승차 인원수")
plt.grid(True)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data) # 헤더를 제외한 데이터를 가져옴
next(data) # 헤더를 제외한 데이터를 가져옴
print(header)

max_station = ""
max_people = 0
result = []
for row in data: 
    for i in range(4, 52): 
        row[i] = row[i].replace(',', "")

    row[4:] = map(int, row[4:]) # 4번째 인덱스부터 52번째 인덱스까지의 데이터를 int로 변환
    if sum(row[10:15:2]) > max_people: # 오전 7~9시 승차인원이 max_people보다 크면
        max_people = sum(row[10:15:2]) # max_people을 오전 7~9시 승차인원으로 변경
        max_station = row[3] + '(' + row[1] + ")" # max_station을 역이름과 지하철호선으로 변경


f.close()

print(f"오전 7~9시 승차 인원이 최대인 역은 {max_station} : {max_people}명")
# result.sort() # 오름차순으로 정렬
# plt.rc('font', family='Malgun Gothic')
# plt.figure(figsize = (10, 3))
# plt.bar(range(len(result)), result)
# plt.xlabel("역")
# plt.ylabel("승차 인원수")
# plt.grid(True)
# plt.show()
```
