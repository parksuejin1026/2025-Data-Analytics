```python
# 파이썬에서 수치 계산, 특히 다차원 배열(array) 연산을 효율적으로 처리하기 위한 핵심 라이브러리인 NumPy를 'np'라는 이름으로 불러옵니다.
import numpy as np

# 데이터 분석 및 조작을 위한 라이브러리인 Pandas를 'pd'라는 이름으로 불러옵니다.
# 엑셀의 표와 같은 형태인 데이터프레임(DataFrame)을 다룰 때 주로 사용됩니다.
import pandas as pd

# 데이터를 그래프나 차트로 시각화하기 위한 Matplotlib 라이브러리의 pyplot 모듈을 'plt'라는 이름으로 불러옵니다.
# 대부분의 그래프는 이 'plt'를 통해 생성하고 수정합니다.
import matplotlib.pyplot as plt

# Matplotlib 라이브러리 전체를 'mpl'이라는 이름으로 불러옵니다.
# 폰트, 스타일 등 그래프의 전반적인 설정을 변경할 때 주로 사용됩니다.
import matplotlib as mpl
```

```python
plt.subplot(2, 2, 1)
plt.plot([1,2,3],[1,2,3])
plt.title("직선")

plt.subplot(2, 2, 2)
plt.plot([1,2,3],[1,4,9])
plt.title("제곱")

plt.subplot(2, 2, 3)
plt.scatter([1,2,3],[1,8,27])
plt.title("세제곱")

plt.subplot(2, 2, 4)
plt.bar([1,2,3],[1,16,81])
plt.title("네제곱")

plt.suptitle("subplot 그래프")
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot([1,2,3],[1,2,3])
plt.yticks([1, 2, 3])
plt.title("선 그래프")

plt.subplot(2, 2, 2)
plt.bar([1,2,3],[2.5, 2, 1.5])
plt.yticks([0, 2])
plt.title("막대 그래프")

plt.subplot(2, 2, 3)
plt.scatter([1, 2, 3],[3, 2, 1])
plt.yticks([1, 2, 3])
plt.title("산점도 그래프")

plt.subplot(2, 2, 4)
plt.hist([1,2,2,2,3,3] ,  bins=[.5, 1.5, 2.5, 3.5], rwidth=.98)
plt.yticks([0, 2])
plt.title("돗수분포 그래프")

plt.suptitle("전체 그래프", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)

x = np.arange(5)
ax[0, 0].plot(x, x)
ax[0, 0].set_title("y = x")

ax[0, 1].plot(x, 2 * x)
ax[0, 1].set_title("y = 2x")

ax[1, 0].plot(x, x ** 2)
ax[1, 0].set_title("y = x ** 2")

ax[1, 1].plot(x, x ** 3)
ax[1, 1].set_title("y = x ** 3")

plt.tight_layout()
plt.show()
```
