### Lecture 18. matplotlib의 scatter 그래프

 __산점도는 가로축과 세로축을 기준으로 두 요소가 서로 어떤 관계를 맺고 있는 지를 파악하기 쉽게 나타낸 그래프__<br>
__예를 들어,__
 1. 가로축을 제주도 지역의 남성 인구수
 2. 세로축을 제주도 지역의 여성 인구수

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter([1,2,3,4], [10,30,20,40], 
            s = [100,200,300,400], # 점 크기 
            c=['red','blue','green','yellow'], # 점 색깔
            marker='s' # 점 모양
            )
plt.grid()
plt.show()
```

```python
plt.scatter([1,2,3,4],[10,30,20,40], #  x축, y축 데이터
            s = [100,200,300,400], # 점 크기
            c = range(4) # 점 색깔 4가지의 색깔을 사용
            )
plt.grid() # 격자 표시
plt.colorbar() # 색상 막대 표시
plt.show() # 그래프 출력
```

```python
plt.scatter([1,2,3,4],[10,30,20,40], #  x축, y축 데이터
            s = [100,200,300,400], # 점 크기
            c = range(4), # 점 색깔 4가지의 색깔을 사용
            cmap='cool' # 색상 지도 설정 (jet, hot, cool, spring 등 다양하게 사용 가능
            )
plt.grid() # 격자 표시
plt.colorbar() # 색상 막대 표시
plt.show() # 그래프 출력
```

```python
plt.style.use('default')
plt.scatter([1,2,3,4],[10,30,20,40], #  x축, y축 데이터
            s = [30,60,90,120], # 점 크기
            c = range(4), # 점 색깔 4가지의 색깔을 사용
            cmap='plasma' # GnBu 색상 지도 설정
            )
plt.colorbar() # 색상 막대 표시
plt.show() # 그래프 출력
```

```python
import random

x = []
y = []
size = []

for i in range(100):
    x.append(random.randint(50,100)) 
    y.append(random.randint(50,100))    
    size.append(random.randint(10,100))

plt.style.use('default') # 스타일 초기화
plt.scatter(x, y,
            s = size,
            c = size,
            cmap = 'jet',
            alpha=0.7 # 투명도 설정 (0~1)
            )
plt.grid()
plt.colorbar()
plt.show()
```

```python
plt.style.use('default')
import numpy as np

# make the data
# np.random.seed(3)

x = 4 + np.random.normal(0, 2, 24) # 평균 0, 표준편차 2, 데이터 개수 24
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()
ax.scatter(x, y,
           s = sizes,
           c= colors,)
ax.set(xlim=(0,8), xticks = np.arange(1, 8),
       ylim=(0,8), yticks = np.arange(1, 8))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots() # 새로운 그래프 영역 생성
for color in ['tab:blue', 'tab:orange', 'tab:green']: # Color strings corrected for standard 'tab:color' format
    n = 75
    x, y = np.random.randn(2, n)  # 2행 75열 난수 생성 x에는 1행 y에는 2행 저장
    scale = 300.0 * np.random.rand(n) # 점 크기 설정
    ax.scatter(x, y, 
               c=color, 
               s=scale, 
               label = color,
               edgecolors='red', # 점 테두리 색깔 설정 
               alpha=0.3) # 반투명도 설정
    
ax.legend()
ax.grid(True)
plt.show()
```
