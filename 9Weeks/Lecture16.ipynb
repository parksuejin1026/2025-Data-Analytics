### Lecture 16. 인구 데이터 준비와 지역의 연령별 인구 데이터 분석

### 우리 동네의 인구 구조를 시각화 하려면 어떤 단계(또는 절차)를 거치면 될까?
1. 인구 데이터 파일을 읽어온다.
2. 전체 데이터에서 한 줄씩 반복해서 읽어온다.
3. 우리 동네에 대한 데이터인지 확인한다.
4. 우리 동네일 경우 0세부터 100세 이상까지의 인구수를 순서대로 저장한다.
5. 저장된 연령별 인구수 데이터를 시각화한다.

```python
## 우리 동네 인구 구조 시각화하기
import matplotlib.pyplot as plt
import numpy as np
import csv

f = open('age.csv', encoding = 'cp949')
data = csv.reader(f)

header = next(data) # 첫 번째 줄의 헤더 추출
print(header)

result = []
for row in data:
    if '신도림' in row[0]:
        for i in row[3:]: # 3번째 칼럼부터 인구 데이터 추출
            result.append(int(i))
        

f.close()
# print(result)

plt.plot(result)
plt.xlabel('Age')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
f = open('age.csv', encoding='cp949')
data = csv.reader(f) # csv 파일 읽기 
header = next(data) # 첫 번째 줄의 헤더 추출

result = []
name = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력하세요 : ")
for row in data:
    if name in row[0]: # 사용자가 입력한 지역 찾기
        for i in row[3:]: # 0세부터 인구 데이터 추출
            result.append(int(i))

f.close()

plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

plt.plot(result)
plt.title(f'{name} 지역의 연령에 따른 인구수')
plt.xlabel('연령')
plt.ylabel('인구수')
plt.grid(True)
plt.show()
```
